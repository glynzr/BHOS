# Matplotlib - Bar Plot

### A bar chart or bar graph is a chart that represents categorical data. The bars can be plotted vertically or horizontally. A bar graph shows comparisons among discrete categories that can include reveniues, prices or other types of data.

### We are going to need `numpy` and `matplotlib.pyplot` libraries.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Creating simple bar chart by using 
#### `plt.figure().add_axes()`</blockquote> function
#### `figure((figsize=(xsize, ysize))`</blockquote>
#### `add_axes([x0, y0, width, height])`</blockquote>

```python
fig = plt.figure(figsize=(8,3))
ax = fig.add_axes([0,0,1,1])

majors = ['IT','Process Automation','Chemical Engineering','Mathematics']
students = [70,120,200,95]

#labeling x values
x = np.arange(len(majors))
plt.xticks(x,majors)

ax.bar(x,students)
plt.title('Students chart')
plt.show()
```

## <blockquote>using `.ylabel()`, `.xlabel()` and changing width, color and fontsizes </blockquote>

```python
fig = plt.figure(figsize=(8,3))
ax = fig.add_axes([0,0,1,1])

majors = ['IT','Process Automation','Chemical Engineering','Mathematics']
students = [70,120,200,95]

x = np.arange(len(majors))
plt.xticks(x,majors)

ax.bar(x,students,label='Students statistics', width=0.5, color="yellow") #label, width and color changed

plt.ylabel("Number of students",fontsize=15)
plt.xlabel("Majors",fontsize=20)
plt.title('Number of students in different majors',fontsize=20)
plt.show()
```

## <blockquote>Comparing multiple data in one chart and labeling them using `.legend()` </blockquote>

```python
years = [2018,2019,2020,2021]

produced_products = [70,150,95,200] #first data
imported_products = [25,100,70,100] #second data
sold_products = [55,145,60,170] #third data

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

x = np.arange(len(years))

plt.title("Products statistics",fontsize=25)
plt.xticks(x,years)
plt.xlabel('years')
plt.ylabel('number of products')

ax.bar(x - 0.25, produced_products, color = 'yellow', width = 0.25, label='produced')
ax.bar(x, imported_products, color = 'lightgreen', width = 0.25,label='imported')
ax.bar(x + 0.25, sold_products, color = 'purple', width = 0.25, label='sold')
plt.legend() #we show which color represents which products by their labels
plt.show()
```

## <blockquote>Showing data on top of each other by using `bottom`</blockquote> 

```python
classes = ['9th class','10th class','11th class']

data = [[19,20,22],[20,18,24]]

fig = plt.figure(figsize=(4,8))
ax = fig.add_axes([0,0,1,1])

x = np.arange(len(classes))
plt.xticks(x,classes)
plt.yticks(np.arange(0,50,2)) #we can change how the y -values distributed

plt.title("Football or basketball?",fontsize=22)
plt.xlabel('classes',fontsize=15)
plt.ylabel('number of students',fontsize=15)

#notice that x positions are the same
ax.bar(x, data[0], color = 'green', width = 0.6, label='footbal')
ax.bar(x, data[1], color = 'orange', width = 0.6,bottom=data[0],label='basketball') #this bar will no longer start from zero because of bottom is defined
                                                                                    
plt.legend(fontsize=15, loc= (1.05, 0.8 ))
plt.show()
```
