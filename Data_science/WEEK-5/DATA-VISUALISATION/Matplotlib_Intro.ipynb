# Matplotlib introduction

To make necessary statistical inferences, it becomes essential to visualize your data, and Matplotlib is one such solution for Python users. It is a comprehensive plotting library useful for those working with Python and NumPy.

    - Matplotlib is an open-source drawing library that supports various drawing types
    - You can generate plots, histograms, bar charts, and other types of charts with  just a few lines of code
    - Itâ€™s often used in web application servers, shells, and Python scripts

Pyplot is a Matplotlib module that provides simple functions for adding plot elements, such as lines, images, text, etc. to the axes in the current figure.

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Let's begin our lesson with a simple graph that uses fundamental Matplotlib code in Jupyter Notebook.

```python
# plotting to our canvas
plt.plot([1, 2, 3], [1, 2, 3])

# showing what we plotted
plt.show()
```

Note that the first array appears on the x-axis, and the second array appears on the y-axis of the plot.

Let us now see how we can add a title, as well as the x-axis and y-axis names, using the title(), xlabel(), and ylabel() methods, respectively.

```python
x = [4, 2, 3]
y = [2, 10, 3]
plt.plot(x, y)
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
```

We can also specify the size of the figure using the figure() method. Additionally, we can pass values as tuples, which make up the length of rows and columns to the argument figsize.

```python
plt.plot([1, 2, 3], [4, 5, 2])
plt.figure(figsize=(10, 5), dpi=50)
```

## Matplotlib Subplots

You can use the subplot() method to add more than one plot in a figure. 

<blockquote>Syntax: plt.subplots(nrows, ncols, index)</blockquote>

The three-integer arguments specify the number of rows and columns and the index of the subplot grid.

Example:

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [3, 2, 1])
plt.title("1st subplot")

plt.subplot(1, 2, 2)
plt.plot([2, 2, 3], [3, 2, 1], 'r')
plt.title("2nd subplot")

plt.show()
```

The add_subplot() function of the figure class enables us to add a graph inside a graph.

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.plot([1, 3, 9])

ax2 = fig.add_subplot(2, 2, 1) # change to (4, 4, 1)
ax2.plot([1, 2, 1])
```

NOTE : Different method. Drawing subgraphs inside another graph and controlling positions!

```python
x = np.linspace(0, 10, 25)
y = x**2 + 2
fig = plt.figure(figsize=(10, 5), dpi=60)

# Control the left, right, width and height of the canvas (from 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # big axes
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3]) # inside, small canvas

# we need a data
axes1.plot(x, y, 'r')
axes2.plot(y, x, 'g')
```

<blockquote>Example : Displaying 2 or more different graphs with legends!

```python
fig, axes = plt.subplots()

axes.set_xlabel("x-label")
axes.set_ylabel("y-label")
axes.set_title("Title of the graphics!")

axes.plot(x, x)
axes.plot(x, x**2, lw=4, linestyle='-', marker='o')
axes.plot(x, x**3, color="red", alpha=0.5, linestyle=':')
# linestyle: '-', '-.', ':'
# markers: 'o', '+', 's', 'l'
# markersize: 5
# markerfacecolor='green'

axes.legend(["y = x", "y = x**2", "y = x**3"], loc=6) # change position: loc = 1 .. or ... 4
```
